```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL da página
url = "https://duome.eu/vocabulary/pt/en"

# Fazendo a requisição
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    words = []
    
    # Encontra todos os itens <li>
    for li in soup.find_all('li'):
        # Captura a palavra em inglês
        speak_span = li.find('span', class_='speak xs voice')
        english_word = speak_span.text.strip() if speak_span else None

        # Captura a tradução em português
        traducao_small = li.find('small', class_='traduccion cCCC wT')
        portuguese_translation = traducao_small.text.strip() if traducao_small else None

        if english_word and portuguese_translation:
            words.append({
                'english': english_word,
                'portuguese': portuguese_translation
            })

    # Cria DataFrame
    df = pd.DataFrame(words)

    # Salva em CSV
    df.to_csv('duome_vocabulary.csv', index=False, encoding='utf-8-sig')
    
    print("Arquivo 'duome_vocabulary.csv' criado com sucesso!")
else:
    print(f"Erro ao acessar a página: {response.status_code}")
```
