```python
#


list_words = ["red",
"yellow",
"green",
"dark ",
"pale green",
"blue",
"purple",
"pink",
"white ",
"gray",
"black ",
"brown",
"orange",
"cream ",
"navy blue",
"plastic",
"wood",
"paper",
"wool",
"leather",
"metal",
"glass",
"fabric",
"denim",
"silk",
]
```

```python
import cv2
import os
import numpy as np

# Caminhos
path_images = "data/raw_selected"
output_folder = "data/recortes"
os.makedirs(output_folder, exist_ok=True)

# Lista de imagens
imagens_paths = sorted([
    os.path.join(path_images, f)
    for f in os.listdir(path_images)
    if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))
])

# Hist√≥rico por imagem
historico = {}
i = 0

while i < len(imagens_paths):
    path = imagens_paths[i]
    img = cv2.imread(path)

    if img is None:
        print(f"‚ö†Ô∏è Erro ao carregar {path}. Pulando...")
        i += 1
        continue

    nome_base = os.path.splitext(os.path.basename(path))[0]
    historico[nome_base] = []

    print(f"\nüñºÔ∏è Imagem {i + 1}/{len(imagens_paths)}: {path}")

    while True:
        display = img.copy()
        cv2.putText(display, f"{nome_base}", (10, 25),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

        img = cv2.resize(img, (640, 700))
        cv2.imshow("Imagem", img)
        key = cv2.waitKey(0)

        if key == ord('r'):
            r = cv2.selectROI("Imagem", img, fromCenter=False, showCrosshair=True)
            if r != (0, 0, 0, 0):
                x, y, w, h = r
                recorte = img[y:y+h, x:x+w]
                nome_base = list_words.pop(0)
                nome_arquivo = f"{nome_base}.png"
                caminho_salvar = os.path.join(output_folder, nome_arquivo)
                cv2.imwrite(caminho_salvar, recorte)
                print(f"‚úÖ Recorte salvo: {nome_arquivo}")
                # historico[nome_base].append(caminho_salvar)

        elif key == ord('n'):
            i += 1
            break

        elif key == ord('p') and historico[nome_base]:
            ultimo = historico[nome_base].pop()
            if os.path.exists(ultimo):
                os.remove(ultimo)
                print(f"‚Ü©Ô∏è Recorte desfeito: {os.path.basename(ultimo)}")
            else:
                print("‚ùå N√£o foi poss√≠vel desfazer o recorte.")

        elif key == ord('q'):
            print("üõë Encerrando...")
            imagens_paths = 0
            cv2.destroyAllWindows()
            exit()

cv2.destroyAllWindows()
```
