```python
import tkinter as tk
from tkinter import messagebox
from deep_translator import GoogleTranslator


def translate_sentence(sentence, source_lang="en", target_lang="pt"):
    """Traduz uma frase usando GoogleTranslator."""
    try:
        translator = GoogleTranslator(source=source_lang, target=target_lang)
        return translator.translate(sentence)
    except Exception as e:
        print(f"Erro ao traduzir a frase '{sentence}': {e}")
        return "Erro na tradução"


def update_translation(text_widget, translation_widget):
    """Atualiza o campo de tradução com base no texto atual."""
    sentence = text_widget.get("1.0", tk.END).strip()
    translation = translate_sentence(sentence)
    translation_widget.delete("1.0", tk.END)
    translation_widget.insert(tk.END, translation)


def show_sentence(index, sentences, widgets, result_list, window):
    """Exibe a sentença atual e configura os botões."""
    if index >= len(sentences):
        messagebox.showinfo("Fim", "Não há mais sentenças para exibir.")
        window.quit()
        return

    sentence = sentences[index]
    widgets["text"].delete("1.0", tk.END)
    widgets["translate"].delete("1.0", tk.END)
    widgets["text"].insert(tk.END, sentence)

    def save():
        result_list.append(sentence)
        show_sentence(index + 1, sentences, widgets, result_list, window)

    def skip():
        show_sentence(index + 1, sentences, widgets, result_list, window)

    widgets["save"].config(command=save)
    widgets["skip"].config(command=skip)


def main(text):
    """Função principal que inicializa a interface e gerencia o fluxo."""
    sentences = [s.strip() for s in text.split(".\n") if s.strip()]
    result_list = []

    window = tk.Tk()
    window.title("Tradutor de Frases")
    window.geometry("800x300")

    # Layout
    text_frame = tk.Frame(window)
    text_frame.pack(fill="both", expand=True, padx=10, pady=10)

    text_widget = tk.Text(text_frame, height=5)
    text_widget.pack(side="left", fill="both", expand=True, padx=(0, 5))

    translate_widget = tk.Text(text_frame, height=5)
    translate_widget.pack(side="right", fill="both", expand=True, padx=(5, 0))

    # Botões
    button_frame = tk.Frame(window)
    button_frame.pack(pady=10)

    save_button = tk.Button(button_frame, text="Salvar")
    skip_button = tk.Button(button_frame, text="Pular")
    translate_button = tk.Button(button_frame, text="Traduzir",
                                 command=lambda: update_translation(text_widget, translate_widget))

    save_button.pack(side="left", padx=10)
    skip_button.pack(side="left", padx=10)
    translate_button.pack(side="left", padx=10)

    widgets = {
        "text": text_widget,
        "translate": translate_widget,
        "save": save_button,
        "skip": skip_button
    }

    show_sentence(0, sentences, widgets, result_list, window)
    window.mainloop()
    return result_list


if __name__ == "__main__":
    with open("database/txt/text.txt", "r", encoding="utf-8") as file:
        text_content = file.read()

    saved = main(text_content)
    print("Sentenças salvas:", saved)
```

```python
import tkinter as tk
from tkinter import messagebox
from deep_translator import GoogleTranslator


def translate_sentence(sentence, source_lang="en", target_lang="pt"):
    try:
        translator = GoogleTranslator(source=source_lang, target=target_lang)
        return translator.translate(sentence)
    except Exception as e:
        print(f"Erro ao traduzir a frase '{sentence}': {e}")
        return "Erro na tradução"


def update_translation(text_widget, translation_label):
    """Atualiza a tradução abaixo do texto."""
    sentence = text_widget.get("1.0", tk.END).strip()
    translation = translate_sentence(sentence)
    translation_label.config(text=translation)


def display_sentence(index, sentences, text_widget, translate_label, result_list, save_button, skip_button, window):
    sentence = sentences[index]
    text_widget.delete("1.0", tk.END)
    text_widget.insert(tk.END, sentence)
    translate_label.config(text="")  # limpa a tradução anterior

    def save_sentence():
        sentences_in_widget = text_widget.get("1.0", tk.END).strip()
        result_list.append(sentences_in_widget)

        # print(translate_label)
        next_sentence()

    def next_sentence():
        if index + 1 < len(sentences):
            display_sentence(index + 1, sentences, text_widget, translate_label, result_list,
                             save_button, skip_button, window)
        else:
            messagebox.showinfo("Fim", "Não há mais sentenças para exibir.")
            window.quit()

    save_button.config(command=save_sentence)
    skip_button.config(command=next_sentence)


def main(text):
    sentences = [s.strip() for s in text.split(".\n") if s.strip()]
    result_list = []

    window = tk.Tk()
    window.title("Tradutor de Frases")
    window.geometry("800x350")

    # Frame principal com o texto original
    text_frame = tk.Frame(window)
    text_frame.pack(fill="both", expand=True, padx=10, pady=10)

    text_widget = tk.Text(text_frame, height=5, font=("Arial", 12))
    text_widget.pack(fill="both", expand=True)
    
    # Label de tradução com fonte menor
    translate_label = tk.Label(window, text="", font=("Arial", 10), fg="gray", wraplength=750, justify="left")
    translate_label.pack(pady=(0, 10))

    # Botões
    button_frame = tk.Frame(window)
    button_frame.pack(pady=10)

    save_button = tk.Button(button_frame, text="Salvar")
    skip_button = tk.Button(button_frame, text="Pular")
    translate_button = tk.Button(button_frame, text="Traduzir",
                                 command=lambda: update_translation(text_widget, translate_label))

    save_button.pack(side="left", padx=10)
    skip_button.pack(side="left", padx=10)
    translate_button.pack(side="left", padx=10)

    # Começar a mostrar as frases
    display_sentence(0, sentences, text_widget, translate_label, result_list, save_button, skip_button, window)

    window.mainloop()
    return result_list


if __name__ == "__main__":
    with open("database/txt/text.txt", 'r', encoding='utf-8') as f:
        text = f.read()

    saved_sentences = main(text)
    print("Sentenças salvas:", saved_sentences)
```
